```python
import torch
from transformers import AutoModelForSemanticSegmentation
from safetensors.torch import load_file
```

```python
model_path = "../sidewalk_segmentation/sidewalk_segmentation_model/model.safetensors"
model_dir = "../sidewalk_segmentation/sidewalk_segmentation_model"
```

```python
model_state = load_file(model_path)
model = AutoModelForSemanticSegmentation.from_pretrained(model_dir)
model.load_state_dict(model_state)
model.eval()
```

```python
dummy_input = torch.randn(1, 3, 512, 512)

torch.onnx.export(
    model,
    dummy_input,
    f"{model_dir}/model.onnx",
    export_params=True,
    opset_version=11,
    input_names=['pixel_values'],
    output_names=['logits']
)

print("Done")
```
